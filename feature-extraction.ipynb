```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df
# batting team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five 
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts()>600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['current_score'] = df.groupby(['match_id'])['runs'].cumsum()
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df = final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

***XG-BOOST***

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.metrics import accuracy_score, classification_report
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred1 = pipe.predict(X_test)
```

```python
print(r2_score(y_test,y_pred1))
print(mean_absolute_error(y_test,y_pred1))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```

***GRADIENT BOOSTING***

```python
from sklearn.ensemble import GradientBoostingClassifier
from tqdm import tqdm
```

```python
gradient_boosting_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
```

```python
def update_progress_bar(estimator, start):
    pbar.n = start + 1  # Update the progress bar position
    pbar.update(1)  # Increment the progress bar


with tqdm(total=100) as pbar:
    gradient_boosting_model.fit(X_train, y_train, callbacks=[lambda start: update_progress_bar(gradient_boosting_model, start)])
```

```python
y_pred9 = gradient_boosting_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred9)
print(f"Accuracy: {accuracy}")
```

```python
print(r2_score(y_test,y_pred9))
print(mean_absolute_error(y_test,y_pred9))
```

```python
print(classification_report(y_test, y_pred9))
```
